# Vanishing Gradient and Exploding Gradient



## The Vanishing Gradient Problem

What it is:

As you backpropagate from the output layer toward the input layer, the gradient gets multiplied by the weights and the derivative of the activation functions at each layer. If these values are small (less than 1), the gradient shrinks exponentially. By the time it reaches the early layers, it is practically zero.

Why it happens:

Deep Networks: The more layers the signal has to travel through, the more multiplications occur.

Activation Functions: Traditional functions like Sigmoid or Tanh "squash" input into a small range. The maximum derivative of Sigmoid is only 0.25.
+1

The Result:
The weights in the early layers of the network never get updated. The "brain" of the model stops learning, and the network fails to converge.

The Exploding Gradient Problem

What it is:
This is the opposite of vanishing. If the weights are large (greater than 1), the repeated multiplication causes the gradient to grow exponentially as it moves backward through the layers.

Why it happens:

Large Weight Initialization: If weights are initialized too high.

Recurrent Neural Networks (RNNs): Because RNNs reuse the same weight matrix over many time steps, a weight slightly larger than 1 can quickly snowball into infinity.

The Result:
The model's weights make massive, chaotic jumps during updates. The loss function fluctuates wildly or returns NaN (Not a Number), effectively breaking the model.

How to Fix Them

Fortunately, we’ve developed several "engineering hacks" to keep gradients stable.


Solutions for Vanishing Gradients

- ReLU Activation: Switching from Sigmoid to ReLU (Rectified Linear Unit) helps because its derivative is 1 for all positive inputs, meaning it doesn't "shrink" the signal.

- Residual Connections (ResNets): These provide a "highway" for the gradient to skip layers, allowing the signal to reach earlier layers without being degraded.

- Batch Normalization: This normalizes the inputs to each layer, ensuring they stay in a range where the activation functions are most responsive.

Solutions for Exploding Gradients

- Gradient Clipping: This is a safety "circuit breaker." If the gradient exceeds a certain threshold (e.g., 1.0), it is forcibly scaled down to stay within a manageable range.

- Weight Regularization: Techniques like $L_1$ or $L_2$ regularization penalize large weights, keeping them small.

- Better Initialization: Using techniques like He Initialization or Glorot (Xavier) Initialization sets the starting weights based on the number of neurons, preventing them from being too large or too small at the start.
