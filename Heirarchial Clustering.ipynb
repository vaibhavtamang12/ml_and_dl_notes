# Heirarchial Clustering


Hierarchical clustering is a method of cluster analysis which seeks to build a hierarchy of clusters. Unlike K-Means, which requires you to pre-specify the number of clusters ($k$), hierarchical clustering creates a tree-like structure that shows the relationships between all data points.

What are the Types?

There are two primary "directions" for building the hierarchy:

Agglomerative (Bottom-Up): The most common approach. Every data point starts as its own cluster. In each step, the two "closest" clusters are merged until only one giant cluster remains.

Divisive (Top-Down): You start with one giant cluster containing all data and recursively split it into smaller clusters until every point is its own cluster

Why and When is it Used?

You should reach for hierarchical clustering when:

The "Number of Clusters" is Unknown: You don't want to guess how many groups exist; you want the data to tell you.

Visualizing Relationships is Key: You need to see the "taxonomy" or genealogy of your data.

The Dataset is Small to Medium: Because it calculates distances between all pairs, it can be slow on massive datasets.

Where is it Used? (Applications)

- Bioinformatics: Creating "phylogenetic trees" to show evolutionary links between species based on DNA.

- Marketing: Segmenting customers into a nested hierarchy (e.g., "Broad Shoppers" $\rightarrow$ "High-Spend Tech Lovers").

- Document Organization: Grouping news articles into broad themes and then specific sub-topics.

How is it Used?

The process relies on two main mathematical decisions:

Distance Metric: How do we measure the distance between two points? (e.g., Euclidean distance $d = \sqrt{\sum (x_i - y_i)^2}$).

Linkage Criterion: How do we measure the distance between two clusters?

- Single Linkage: Distance between the closest members.

- Complete Linkage: Distance between the furthest members.

- Ward’s Method: Minimizes the variance within clusters.

The final result is visualized using a Dendrogram—a diagram that looks like an upside-down tree. You "cut" the tree horizontally at a certain height to decide how many clusters you want.

Pros

- Discovery-Oriented: You don't need to know the number of clusters ($k$) in advance. You can decide the optimal number by looking at the results.

- Rich Visualization: The Dendrogram (the tree diagram) provides a clear, intuitive map of how data points relate to one another at different levels of granularity.

- Captures Nested Structures: It is uniquely suited for data that has a natural hierarchy, such as biological taxonomies or corporate organizational structures.

- Flexibility: You can use any distance metric (Euclidean, Manhattan, Cosine) and various linkage methods to fine-tune how clusters are formed.

Cons

- High Computational Cost: It is much slower than K-Means. Because it calculates distances between all pairs of points, it becomes inefficient as your dataset grows into the tens of thousands.

- Irreversibility: Once the algorithm merges two clusters (in the bottom-up approach), that step cannot be undone. If a mistake is made early on, it propagates through the rest of the hierarchy.

- Sensitivity to Noise: Outliers can create "chains" (especially in Single Linkage), leading to clusters that are stretched out rather than compact and distinct.

- Arbitrary "Cuts": While you don't have to pick $k$ at the start, you still have to decide where to "cut" the dendrogram to get your final groups, which can be subjective.
