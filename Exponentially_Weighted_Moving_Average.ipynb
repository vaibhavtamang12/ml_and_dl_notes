# Exponentially Weighted Moving Average

What is Exponentially Weighted Moving Average (EWMA)?

An Exponentially Weighted Moving Average (EWMA) is a statistical measure used to analyze time-series data. Unlike a Simple Moving Average (SMA), which gives equal weight to all data points in a window, EWMA applies weights that decrease exponentially as the data points get older.

In short: Recent data has a much higher impact on the average than older data.

When and Where is it Used?

- EWMA is the go-to tool when you need a "reactive" trend line that doesn't lag as much as a standard average.

- Finance & Trading: Used to track stock prices and volatility. The MACD (Moving Average Convergence Divergence) indicator is built using EWMAs.

- Deep Learning & AI: Used in optimization algorithms like Adam or RMSProp to smooth out gradients during training, preventing the model from overshooting.

- Quality Control: In manufacturing, EWMA charts are used to detect small shifts in a process mean more quickly than standard control charts.

- Network Traffic: To estimate round-trip times (RTT) for data packets to manage congestion.

Why is it Used?

The primary reason to use EWMA is to reduce lag. In a rapidly changing environment (like a volatile stock market or a training neural network), a simple average is often "too slow" because it's still being weighted down by what happened 20 days ago. EWMA stays "closer" to the current price or value while still smoothing out the "noise" (random spikes).

How is it used

The EWMA is calculated recursively. Each new estimate depends on the previous estimate and the current observed value. The mathematical formula is:$$V_t = \beta \cdot V_{t-1} + (1 - \beta) \cdot \theta_t$$

Where:

- $V_t$: The current EWMA value.

- $V_{t-1}$: The previous day's EWMA value.

- $\theta_t$: The current observed value.

- $\beta$: The smoothing weight (usually between 0.9 and 0.99). A higher $\beta$ makes the average smoother and less reactive; a lower $\beta$ makes it jumpy and more reactive to new data.

Pros and Cons

Pros

- Reduced Lag: It reacts to recent trends much faster than a Simple Moving Average.

- Memory Efficiency: You don't need to store a huge window of past data (like the last 100 days). You only need the previous day's average and the current value to calculate the next step.

- Flexibility: By adjusting the $\beta$ (or "alpha") parameter, you can perfectly tune how much you want to prioritize recent data versus historical stability.

- Smoothness: It filters out short-term fluctuations (noise) while highlighting the underlying trend.

Cons

- Complexity: It is harder to calculate and explain to non-technical stakeholders compared to a simple "sum and divide" average.

- Initial Bias: Because it is recursive, the very first few values can be inaccurate because there isn't enough "history" built into the average yet (often solved with bias correction).

- Sensitivity to Outliers: While it smooths noise, a single massive spike in recent data will pull the EWMA much more aggressively than it would an SMA.
