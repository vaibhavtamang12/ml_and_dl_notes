# RNN


A Recurrent Neural Network (RNN) is a class of artificial neural networks designed to process sequential data. Unlike traditional feed-forward neural networks, which assume all inputs are independent of each other, RNNs have "memory." They take information from prior steps in a sequence to influence the current input and output.

RNNs are used when data is sequential or time-dependent.

Common applications:
- 🗣️ Speech recognition (e.g., voice assistants)
- 🌐 Language translation
- ✍️ Text generation / autocomplete
- 💬 Chatbots
- 📈 Stock price prediction
- 🧬 DNA sequence analysis
- 🎬 Video analysis (frame sequences)

👉 Anywhere you need to understand patterns over time or sequence.

Why is it used

RNNs are used because they can:

Capture temporal dependencies (past affects future)
Work with variable-length inputs (sentences, time series)
Maintain a hidden state (memory) across inputs

👉 Example: Understanding the sentence

“He went to the bank to deposit money”
The word bank is understood using previous words.

How is it used

Basic workflow:

Input sequence is fed one step at a time
At each step:
Current input + previous hidden state → new hidden state
Output is generated at each step or at the end

👉 Mathematically:

Hidden state = function(previous state, current input)

You typically implement RNNs using frameworks like:

TensorFlow

PyTorch


Pros of RNN (Advantages)

✔️ Handles sequential data naturally

✔️ Remembers previous information (context)

✔️ Works with variable-length inputs

✔️ Good for time-series prediction

✔️ Parameter sharing reduces complexity

✔️ Can model temporal dynamics

Cons of RNN (Disadvantages)

❌ Vanishing gradient problem (hard to learn long-term dependencies)

❌ Training can be slow and difficult

❌ Cannot easily capture long-range dependencies

❌ Sequential processing → not easily parallelizable

❌ Performance degrades with long sequences

❌ Requires improvements like:

  Long Short-Term Memory (LSTM)
  
  Gated Recurrent Unit (GRU)
