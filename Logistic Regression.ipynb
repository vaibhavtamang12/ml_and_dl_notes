# Logistic Regression

Despite its name, Logistic Regression is a tool for classification, not traditional regression. While linear regression tries to predict a specific number (like a house price), logistic regression predicts the probability of something belonging to a specific category (like "Spam" vs. "Not Spam").

1. The Intuition

Imagine you are trying to predict if a student passes an exam based on how many hours they studied.

- The Linear Problem: If you use a straight line (Linear Regression), your "prediction" might eventually say a student has a 150% chance of passing if they study for 100 hours. That doesn't make sense; probability must be between 0 and 1.

- The Logistic Solution: Logistic regression takes that straight line and "squashes" it into an S-shape (called a Sigmoid function). This curve stays strictly between 0 and 1.

How it works:

- It calculates a "score" using a linear formula ($z = \beta_0 + \beta_1x$).

- It passes that score through the Sigmoid function: $\sigma(z) = \frac{1}{1 + e^{-z}}$.

- The result is a probability (e.g., 0.85).

- You set a threshold (usually 0.5). If the probability is above 0.5, you classify it as "Class 1" (Pass); if below, "Class 0" (Fail).

## When and Why is it used?

- When: Use it when your target variable is categorical (Yes/No, True/False, Buy/Don't Buy).

- Why: It is used because it provides more than just a label—it gives you the certainty of the prediction. Knowing a patient has a 95% risk of a disease is very different from knowing they have a 51% risk, even if both are classified as "High Risk."

Common Use Cases:

- Finance: Predicting if a loan applicant will default.

- Healthcare: Predicting the likelihood of a patient having a specific condition based on symptoms.

- Marketing: Predicting if a user will click on an ad (Click-Through Rate).

The Pros (Advantages)

- Simplicity and Speed: It is one of the fastest algorithms to train and run, making it ideal for systems that need to make real-time predictions.

- Interpretability: Unlike "black box" models (like Deep Learning), you can easily see which factors are the most important. For example, you can tell exactly how much an increase in "Age" affects the probability of a "Purchase."

- Probability Estimates: It doesn't just give you a "Yes" or "No"; it tells you the confidence level (e.g., "There is a 72% chance this is spam").

- Low Resource Usage: It requires very little memory and processing power, meaning it can run on basic hardware or even within mobile apps easily.

- Resistance to Overfitting: In simple datasets, it is less likely to "over-memorize" the data compared to complex models, especially when using regularization techniques (like L1 or L2).

The Cons (Disadvantages)

- Assumes Linearity: It assumes that the relationship between your inputs and the outcome follows a straight line (before being squashed by the S-curve). If your data has complex, wavy patterns, this model will miss them.

- Requires Independent Features: It performs poorly if your input variables are highly correlated with each other (a problem called Multicollinearity).

- Limited Complexity: It is strictly a linear classifier. It cannot solve non-linear problems without manually adding complex math (like polynomial features) to the data beforehand.

- Vulnerability to Outliers: Because it relies on finding a "best fit," a few extremely weird data points can skew the entire S-curve, leading to poor predictions.

- Large Sample Size: While simple, it often needs a relatively large amount of data to achieve high stable accuracy compared

## Classification Metrics

Classification metrics are the scorecards we use to judge how well a machine learning model is performing. When a model predicts a category (e.g., "Sick" vs. "Healthy"), we need more than just a gut feeling to know if it's working; we need mathematical proof.

The Intuition

Imagine a security alarm.

- If it rings when a burglar enters, it's doing its job.

- If it rings because a cat walked by, that’s a False Positive.

- If a burglar enters and it doesn't ring, that’s a False Negative.

Classification metrics allow us to count these specific successes and failures to see if the "alarm" is tuned correctly for the situation.

When and Why are they used?

- When: They are used during the evaluation phase of building a model and after the model is deployed to monitor its health.

- Why: You cannot rely on "Accuracy" alone. For example, if 99% of people are healthy, a model that simply guesses "Healthy" every time will be 99% accurate but 0% useful at finding sick people. Metrics help you find the trade-off between being too sensitive or too cautious.

How are they used? (The Metrics)

To use these metrics, we first create a Confusion Matrix, which is a table showing the breakdown of correct and incorrect predictions.

From this table, we calculate the four heavy hitters:

Accuracy: The percentage of total guesses that were correct.

Best for: When your classes are balanced (50/50).

Precision: "Of all the times the model predicted 'Yes', how often was it actually right?"

Best for: When the cost of a False Positive is high (e.g., calling an innocent person a criminal).

Recall (Sensitivity): "Of all the 'Yes' cases that exist, how many did the model actually find?"

Best for: When the cost of a False Negative is high (e.g., missing a cancer diagnosis).

F1-Score: The harmonic mean of Precision and Recall.

Best for: When you want a balance between Precision and Recall and your data is imbalanced.

To calculate these metrics, we first need to define the four building blocks found in a Confusion Matrix:

- TP (True Positive): Predicted "Yes", and it was "Yes".

- TN (True Negative): Predicted "No", and it was "No".

- FP (False Positive): Predicted "Yes", but it was "No" (Type I Error).

- FN (False Negative): Predicted "No", but it was "Yes" (Type II Error).

The Formulas

Here is how you mathematically calculate the "Big Four" metrics:

Accuracy: The overall correctness of the model.$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$Precision: The accuracy of the "Positive" predictions.$$\text{Precision} = \frac{TP}{TP + FP}$$Recall (Sensitivity): The ability to find all "Positive" cases.$$\text{Recall} = \frac{TP}{TP + FN}$$F1-Score: The balanced average (harmonic mean) of Precision and Recall.$$\text{F1-Score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}$$

## Pros and cons

Accuracy

- Pros:

  - Intuitive and very easy to communicate to non-technical people.

  - Provides a single "at-a-glance" number for overall performance.

- Cons:

  - The Accuracy Paradox: Highly misleading on imbalanced data (e.g., if 99% of transactions are legit, a model that says "all legit" is 99% accurate but useless).

  - Treats all errors (FP and FN) as equally bad, which is rarely true in the real world.

Precision

- Pros:

  - Crucial when the cost of a "False Alarm" is high (e.g., you don't want to mark an important email as Spam).

  - Ensures that when the model says "Yes," you can trust it.

- Cons:

  - Does not care about missed cases (False Negatives). A model could have 100% precision but miss 90% of the actual targets.

Recall

- Pros:

  - Essential when "Missing" a case is dangerous (e.g., diagnosing a disease or detecting a structural crack in a bridge).

  - Focuses on capturing the entire positive population.

- Cons:

  - Often leads to "Over-sensitivity." You might get a high Recall by just predicting "Yes" for everything, which fills your results with False Positives.

F1-Score

- Pros:

  - The best metric for imbalanced datasets because it punishes extreme values of either Precision or Recall.

  - Provides a much more realistic view of model health than Accuracy.

- Cons:

  - Harder to explain to stakeholders ("What does an F1 of 0.76 actually mean?").

  - It treats Precision and Recall as equally important, which might not fit every business case.

## Softmax regression

Softmax Regression (also known as Multinomial Logistic Regression) is essentially the "big brother" of Logistic Regression. While standard Logistic Regression handles binary choices (A vs. B), Softmax Regression handles multiple classes (A vs. B vs. C vs. D).

The Intuition

Imagine you are looking at a handwritten digit and trying to decide if it is a 0, 1, 2, or 3.

- The Score: The model calculates a raw "score" (called a logit) for each possible number.

- The Problem: These scores could be anything (like 10.5, -2.1, 0.5). They don't look like probabilities.

- The Softmax Function: This function takes those raw scores and transforms them so that:

  - Every score is between 0 and 1.

  - All the scores added together equal exactly 1 (or 100%).

When and Why is it used?

- When: Use it for Multi-class Classification where the categories are mutually exclusive (an image can be a cat or a dog, but not both at once).

- Why: Because standard Logistic Regression can only handle two classes. While you could run multiple separate binary models, Softmax is mathematically cleaner and more efficient because it considers all classes simultaneously.

Common Use Cases:

- OCR: Recognizing letters or numbers.

- Image Tagging: Deciding if a photo is of a "Forest," "Beach," or "City."

- NLP: Predicting the next word in a sentence from a vocabulary of thousands of words.

How is it used? (The Formula)

The Softmax function $\sigma$ for a vector of scores $z$ for the $j$-th class is:$$\sigma(z)_j = \frac{e^{z_j}}{\sum_{k=1}^{K} e^{z_k}}$$

The Process:

- Calculate a linear score for every class.

- Exponentiate the scores ($e^z$) to make them all positive.

- Divide each by the sum of all exponentiated scores to get the probability.

- The class with the highest probability is your prediction (the "winner-takes-all").

## Pros and Cons

Pros

- Multi-class native: It naturally handles more than two categories without needing "workarounds."

- Probabilistic output: It gives a clear probability distribution across all classes, showing you the model's second and third choices.

- Differentiable: Because it is a smooth mathematical function, it is perfect for training neural networks using gradient descent.

- Simplicity: It remains a linear model, making it faster to train than complex deep-learning alternatives.

Cons

- Mutually Exclusive Assumption: It assumes an object can only belong to one class. It performs poorly for "multi-label" problems (e.g., a movie that is both "Action" and "Comedy").

- Outlier Sensitivity: Like Logistic Regression, extreme values can heavily influence the exponentiation in the formula.

- Linearity: It cannot capture complex, non-linear relationships between features unless you manually transform the data.

- Computationally heavier: As the number of classes grows (e.g., predicting one of 50,000 words), calculating the denominator (the sum of all scores) becomes very slow.

## Hyperparameters in Logistic Regression

1. Regularization Type (Penalty): This is perhaps the most important hyperparameter. It prevents overfitting (where the model memorizes the training data but fails on new data) by punishing large coefficients.

  - L1 (Lasso): Adds a penalty equal to the absolute value of the magnitude of coefficients. It can force some coefficients to become exactly zero, effectively performing feature selection.
  - L2 (Ridge): Adds a penalty equal to the square of the magnitude of coefficients. It shrinks the weights but keeps them all small, rather than eliminating them.
  - Elastic Net: A middle ground that combines both L1 and L2 penalties.

2. Regularization Strength ($C$): In most libraries (like Scikit-Learn), the hyperparameter $C$ controls the strength of the penalty. Note: $C$ is the inverse of regularization strength.

  - Small $C$ (e.g., 0.01): High regularization. The model is "penalized" heavily for large weights, leading to a simpler, flatter model (underfitting risk).
  - Large $C$ (e.g., 100): Low regularization. The model trusts the training data more and allows for larger weights (overfitting risk).
  
3. Solver: The "solver" is the algorithm used to find the best weights (to minimize the loss function). Since there isn't a single formula to solve logistic regression, we use iterative optimization.

  - liblinear: Good for small datasets and L1 penalty.
  - lbfgs / sag / saga: Better for large datasets. saga is unique because it handles L1, L2, and Elastic Net penalties.
  - newton-cg: Computationally expensive but very precise for multiclass problems
  
4. Multi-class Strategy (multi_class): This dictates how the model handles more than two categories.
  - OvR (One-vs-Rest): It treats a 3-class problem as three separate binary problems (Class A vs. Not A, etc.).
  - Multinomial: It uses the Softmax approach we discussed earlier, calculating probabilities across all classes simultaneously.

5. Maximum Iterations (max_iter): Since solvers are iterative, they need to know when to stop. This sets the maximum number of "steps" the solver takes to find the optimum. If this is too low, the model may stop before it finds the best solution (failing to converge).
