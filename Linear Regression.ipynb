# Linear Regression

Linear regression is one of the most fundamental and widely used algorithms in statistics and machine learning. At its core, it is a method to model the relationship between a dependent variable (what you want to predict) and one or more independent variables (the factors that influence it).

## What is Linear Regression

Linear regression is a "supervised learning" algorithm that finds the best-fitting straight line (known as the regression line) through a set of data points.

Simple Linear Regression: Uses one independent variable to predict a result.Formula: $y = \beta_0 + \beta_1x + \epsilon$

Multiple Linear Regression: Uses two or more independent variables to predict a result.Formula: $y = \beta_0 + \beta_1x_1 + \beta_2x_2 + ... + \beta_nx_n + \epsilon$

Key Components:

$y$: The Dependent Variable (Target).

$x$: The Independent Variable (Predictor).

$\beta_0$: The Y-intercept (the value of $y$ when $x$ is 0).

$\beta_1$: The Slope (how much $y$ changes for every 1-unit change in $x$).

$\epsilon$: The Error term (the difference between actual data and the predicted line).

## Why is it used?

It is used for two main purposes:

- Prediction and Forecasting: To estimate future values. (e.g., "If I spend $\$5,000$ on ads next month, what will my sales be?")

- Determining Relationship Strength: To see how strongly an independent variable affects a dependent one. (e.g., "Does the number of hours studied actually impact test scores significantly?")

## When is it used?

You should use linear regression when:

The relationship between your variables is roughly linear (they move together in a straight-line pattern).

The target variable you are trying to predict is continuous (like price, weight, or temperature), rather than categorical (like "Yes/No").

You need a model that is easy to explain to stakeholders.

## How is it used?

- Data Collection: Gather historical data for both the predictor (X) and the target (Y).

- Model Training: The algorithm uses a method called Ordinary Least Squares (OLS) to minimize the sum of the squares of the vertical deviations between each data point and the line.

- Evaluation: You check the model's accuracy using metrics like R-squared (which tells you what percentage of the data's variation is explained by the model).

- Deployment: Once satisfied, you plug in new "X" values to get a predicted "Y."

## Pros

- Simplicity: Very easy to understand, explain, and implement.

- Efficiency: Requires very little computing power compared to Deep Learning.

- Interpretability: You can clearly see which factors are most important based on the coefficients ($\beta$).

- Fast Training: Works very quickly even on large datasets.

## Cons

- Linearity Assumption: It fails if the relationship is curved or complex.

- Sensitive to Outliers: A single extreme data point can "pull" the line and ruin the accuracy.

- Over-simplification: Real-world data is rarely perfectly linear.

- Independence: Assumes variables aren't related to each other (Multicollinearity), which is often untrue.

## Regression Metrics



1. Absolute Error Metrics

These metrics tell you the average size of the mistake in the same units as your data.

MAE (Mean Absolute Error): The average of the absolute differences between actual and predicted values. It treats all errors equally and is very robust to outliers.

Formula: $MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$

MAPE (Mean Absolute Percentage Error): Measures error as a percentage of the actual values.8 It is useful for communicating accuracy to business stakeholders (e.g., "Our model is 95% accurate").

Constraint: It cannot be used if any actual values are zero.

2. Squared Error Metrics

These metrics square the errors, which means they penalize "big misses" much more heavily than small ones.

MSE (Mean Squared Error): The average of squared differences.Because it's squared, the units are also squared (e.g., if you're predicting price in dollars, MSE is in dollars squared), making it hard to interpret intuitively.

Formula: $MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$

RMSE (Root Mean Squared Error): The square root of MSE.This is the most popular metric because it penalizes large errors (like MSE) but brings the units back to the original scale (like MAE).

Formula: $RMSE = \sqrt{MSE}$

RMSLE (Root Mean Squared Logarithmic Error): Only cares about the relative error.It is used when the target variable has a wide range of values (exponential growth) and you don't want to over-penalize large absolute differences.

3. R-Squared (Goodness of Fit)

These metrics don't measure "error" in units, but rather the strength of the relationship.

$R^2$ (Coefficient of Determination): Tells you what percentage of the data's variation is explained by the model.

$R^2 = 1$ means the model is perfect.

$R^2 = 0$ means the model is no better than just guessing the average value.

Adjusted $R^2$: A modified version of $R^2$ that accounts for the number of predictors.Standard $R^2$ always goes up when you add more variables, even if they are useless. Adjusted $R^2$ will only increase if the new variable actually improves the model.

Which one should you use?

MAE: When you want a simple, robust average error and don't care about outliers.

RMSE: When large errors are particularly bad (e.g., medical dosage, structural safety).

$R^2$: When you want to know how much "information" your model is capturing.

Adjusted $R^2$: Always use this instead of $R^2$ if you have multiple independent variables.

# Gradient Descent

Gradient descent is the engine that allows machine learning models to "learn." While linear regression is the model, gradient descent is the optimization algorithm used to find the best possible version of that mode

## What is Gradient Descent

Gradient descent is an iterative optimization algorithm used to find the minimum value of a function. In machine learning, that function is usually a "Loss Function" (the measure of how wrong the model's predictions are).


The Analogy: Imagine you are standing at the top of a foggy mountain and want to get to the very bottom (the valley). Because of the fog, you can't see the path. However, you can feel the steepness of the ground under your feet.


- You feel which way is downhill (the negative gradient).

- You take a step in that direction (the learning rate).

- You repeat this until you reach the lowest point where the ground is flat.

## Why and when is it used

Why use it?

It is used because most machine learning problems are too complex to solve with a direct formula (like the "Normal Equation" in simple math). For models with thousands or millions of variables (like Deep Learning), calculating the answer directly would take years. Gradient descent finds the answer efficiently by "walking" toward it.

When use it?

- When the dataset is too large to fit into memory for a direct calculation.

- When training complex models like Neural Networks, Logistic Regression, or Support Vector Machines (SVM).

- When you need to minimize a differentiable function (one where you can calculate the slope).

## How is it used

The algorithm follows a simple mathematical loop:

- Initialize: Start with random values for your parameters (weights and biases).

- Calculate Gradient: Use calculus (derivatives) to find the slope of the loss function at the current point.This tells you which direction is "downhill"

- Update Parameters: Move the parameters a small amount in the opposite direction of the gradient.

  - Formula: $\theta_{new} = \theta_{old} - \alpha \cdot \nabla J(\theta)$
  - $\alpha$ (Alpha): This is the Learning Rate. If it's too big, you might overstep the valley.19 If it's too small, the walk will take forever.

- Repeat: Continue until the slope is nearly zero, meaning you've reached the bottom (Convergence).

## Types of Gradient Descent

Batch GD

- Uses the entire dataset

- Very stable; guranteed to converge

- Very slow on large data; high memory usage

Stochastic GD

- Uses only one random data point to calculate gradient for each step

- Extremely fast; can jump out of local pits

- Path is very noisy; never truly settles at the bottom

Mini Batch GD

- Uses a small subset

- Industry standard

- Requires tuning the batch_size parameter
