# AdaDelta

Adadelta is an advanced gradient descent optimization algorithm designed to improve upon AdaGrad. Its primary goal is to address two major drawbacks of its predecessor: the continuous decrease of the learning rate and the need for a manually selected global learning rate.

When and where is it used

Adadelta is used in training deep neural networks, particularly in scenarios where:

- Sparse Data: You are dealing with data where some features appear much less frequently than others (e.g., NLP or recommendation systems).

- Long Training Sessions: You want to avoid the "vanishing learning rate" problem where the model stops learning before reaching the optimum.

- No Hyperparameter Tuning: You don't have the time or resources to find the "perfect" initial learning rate.

Why is it used

To understand why Adadelta exists, you have to look at the flaw in AdaGrad. AdaGrad keeps a running sum of all past squared gradients. Since these are always positive, the sum grows indefinitely, causing the learning rate to shrink until it becomes so small the model effectively stops learning.

Adadelta solves this by:

- Restricting the Window: Instead of summing all past gradients, it uses an exponentially moving average of squared gradients.

- Eliminating the Learning Rate: It uses a ratio of the moving averages of both the gradients and the parameter updates themselves. This means you technically don't even need to set a starting learning rate ($\eta$).

How is it used

In most frameworks like TensorFlow or PyTorch, it is implemented as a standard optimizer class. Mathematically, it replaces the fixed learning rate with a dynamic quantity based on the "units" of the parameters.

The update rule looks like this:$$\Delta \theta_t = - \frac{RMS[\Delta \theta]_{t-1}}{RMS[g]_t} \cdot g_t$$

- $g_t$: The current gradient.

- $RMS[g]_t$: The Root Mean Square of previous gradients (the "decaying" history).

- $RMS[\Delta \theta]_{t-1}$: The Root Mean Square of previous weight updates.

Pros and Cons

Pros

- No Manual Learning Rate: It is one of the few optimizers where you don't need to specify an initial learning rate; it adapts based on the progress of the training.

- Prevents Learning Rate Decay: By using a moving window of gradients, it ensures the learning rate doesn't drop to zero prematurely.

- Robust to Large Gradients: It handles noisy gradients and large jumps in the loss function quite smoothly.

- Computational Efficiency: It requires very little memory beyond storing the moving averages.

Cons

- Slow Convergence: Because it is so cautious, Adadelta can sometimes take much longer to reach the minimum compared to Adam or RMSProp.

- Sensitivity to Decay Factor: While you don't need a learning rate, you do need a "decay rate" ($\rho$), and if this is set poorly, the model might struggle.

- Less Common in Modern SOTA: In current research (like training LLMs), Adam and its variants (AdamW) have largely overshadowed Adadelta due to their superior speed.
