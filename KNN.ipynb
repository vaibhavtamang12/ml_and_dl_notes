# KNN

K-Nearest Neighbors (KNN) is one of the simplest and most intuitive machine learning algorithms. It is a supervised learning algorithm used for both classification (predicting a label) and regression (predicting a continuous value).

The core idea is simple: "Birds of a feather flock together." To predict the value of a new data point, KNN looks at the $k$ most similar data points (the "neighbors") in the training set and makes a prediction based on their values.

## When and where is it used

KNN is a "lazy learner," meaning it doesn't build a complex internal model; it just remembers the training data. It is commonly used in:

Recommendation Systems: Suggesting movies or products based on what similar users liked (e.g., "Users who liked X also liked Y").

Pattern Recognition: Identifying handwriting, facial recognition, or credit scoring.

Medical Diagnosis: Classifying a patient’s risk level based on the profiles of similar past patients.

Anomaly Detection: Finding "outliers" that don't belong to any known cluster (e.g., credit card fraud detection).

## Why is it used

Simplicity: It’s incredibly easy to explain to stakeholders.

No Assumptions: Unlike many algorithms, KNN doesn't assume your data follows a specific distribution (like a straight line or a bell curve). This makes it a non-parametric algorithm.

Versatility: It handles multi-class problems and regression with the same underlying logic

## How is it Used

To use KNN, you generally follow these steps:

- Choose the value of $k$: This is the number of neighbors the algorithm will consult.

- Calculate Distance: When a new data point arrives, the algorithm calculates the distance between it and all other points in the dataset. Common formulas include Euclidean distance:$$d(p, q) = \sqrt{\sum_{i=1}^{n} (p_i - q_i)^2}$$

- Find Neighbors: It selects the $k$ points with the smallest distances.Vote (Classification) or Average (Regression):
  -  For classification, the new point takes the label of the majority.
  
  - For regression, it takes the average value of those $k$ neighbors.

## Pros and Cons

Pros

- Easy to implement: Very few lines of code are required.

- No Training Phase: Since it just stores the data, "training" is instantaneous.

- Naturally handles multi-class: It doesn't struggle with datasets that have more than two categories.


Cons

- Computationally Expensive: Because it has to calculate the distance to every single point every time you make a prediction, it gets very slow as your dataset grows.

- Memory Intensive: You have to store the entire dataset in memory.

- Sensitive to Scale: If one feature (like "Salary") has much larger numbers than another (like "Age"), the larger numbers will dominate the distance calculation. You must normalize or scale your data first.

- The "Curse of Dimensionality": It performs poorly if you have too many features (columns), as the "distance" between points becomes less meaningful.
