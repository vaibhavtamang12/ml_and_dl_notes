# Weight Initialization

In deep learning, the two most prominent weight initialization techniques are Xavier (Glorot) Initialization and He (Kaiming) Initialization. These methods were developed to solve the "vanishing" and "exploding" gradient problems by maintaining a consistent variance of activations across layers.

Xavier Initialization

When and Where it is Used

Xavier initialization is the standard choice for layers using Sigmoid or Hyperbolic Tangent (tanh) activation functions. Because these functions are symmetric and have a gradient that peaks at zero, they require weights that keep the signal within the linear region of the activation.

Why it is Used

If weights are too small, the signal shrinks as it passes through layers (vanishing); if too large, the signal saturates the sigmoid/tanh function, leading to zero gradients. Xavier initialization keeps the variance of the input and output of a layer the same.

How it is Used

Weights are sampled from a distribution with a mean of 0 and a variance $\sigma^2$ defined by:$$\sigma^2 = \frac{2}{n_{in} + n_{out}}$$Where $n_{in}$ is the number of input units and $n_{out}$ is the number of output units.

Pros and Cons

Pros:

- Prevents gradients from vanishing or exploding in networks using tanh/sigmoid.

- Ensures that every layer in the network initially has roughly the same distribution of activations.

- Mathematically grounded for symmetric activations.

Cons

- Performs poorly with Rectified Linear Unit (ReLU) activations, as it doesn't account for the fact that ReLU "shuts off" half the input.

- Can lead to very small gradients in extremely deep networks if not paired with the correct activation.

He(Kaiming Initialization)

When and Where it is Used

He initialization is the go-to technique for networks using ReLU or its variants (Leaky ReLU, ELU). It is the default for most modern architectures like ResNet or specialized CNNs.

Why it is Used

ReLU units output zero for half of their input space on average. Xavier initialization doesn't account for this "halving" of the variance. He initialization compensates by doubling the variance of the weights, ensuring the signal doesn't die out as it moves deeper into the network.

How is it used

Weights are sampled from a distribution with a mean of 0 and a variance $\sigma^2$ defined by:$$\sigma^2 = \frac{2}{n_{in}}$$By only focusing on $n_{in}$, it provides a larger initial weight range to counteract the sparsity of ReLU.

Pros and Cons

Pros

- Essential for training deep networks with ReLU (modern industry standard).

- Allows for much faster convergence in deep architectures.

- Robust against the "Dead ReLU" problem during the initial phases of training.

Cons

- Can cause exploding gradients if used with Sigmoid or Tanh activations because the weights are "too large" for those functions.

- Slightly more prone to initial instability if the learning rate is set too high compared to Xavier.
