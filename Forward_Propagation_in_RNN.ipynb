# Forward Propagation in RNN

Forward propagation in a Recurrent Neural Network (RNN) is the process of moving input data through the network's layers to generate an output or prediction. Unlike standard feed-forward neural networks, RNNs are designed to handle sequential data by maintaining a "memory" of previous inputs.

In an RNN, forward propagation involves calculating the state of a hidden layer and the output at each discrete time step t. The key distinction is the hidden state (ht​), which acts as a summary of all preceding inputs (x1​,x2​,…,xt​).

The mathematical operations at each time step t are:

    Calculate the Hidden State:
$$h_t = \phi(W_{hh}h_{t-1} + W_{xh}x_t + b_h)$$

        xt​: Input at current time step.

        ht−1​: Hidden state from the previous time step.

        Wxh​,Whh​: Weight matrices for input and hidden states.

        ϕ: Activation function (usually tanh or ReLU).

    Calculate the Output:
$$y_t = \psi(W_{hy}h_t + b_y)$$

        yt​: Output at current time step.

        ψ: Activation function (e.g., Softmax for classification).

When and Where is it Used?

Forward propagation is used every time the model makes a prediction or processes a sequence.

    When: It occurs during both the training phase (to calculate the initial error/loss before backpropagation) and the inference phase (when the trained model is used to predict results).

    Where: It is the core mechanism in any application involving temporal or sequential data, such as:

        Natural Language Processing (NLP): Predicting the next word in a sentence.

        Time Series Analysis: Forecasting stock prices or weather patterns.

        Speech Recognition: Converting audio signals into text.

Why is it Used?

Standard neural networks assume all inputs and outputs are independent. However, for many tasks, context matters.

    Sequential Context: If you want to predict the next word in a sentence, you need to know the words that came before it. Forward propagation allows the network to pass information through time.

    Parameter Sharing: In an RNN, the same weights (W) are used at every time step. This allows the model to find patterns regardless of where they appear in the sequence, making it highly efficient for variable-length data.

How is it Used? (The Step-by-Step Flow)

    Initialization: The initial hidden state (h0​) is usually set to zeros.

    Unrolling: The sequence is processed one element at a time. For a sequence of length T, the forward pass "unrolls" into T steps.

    Iteration: At each step t:

        The current input xt​ and previous hidden state ht−1​ are combined.

        The new hidden state ht​ is computed.

        An output yt​ is produced (if the architecture requires an output at every step).

    Loss Calculation: During training, the final output is compared to the actual target to compute the Loss Function, which is then used by "Backpropagation Through Time" (BPTT) to update the weights.
