# Bahdanau Attention

The Bahdanau Attention Mechanism, often called Additive Attention, was introduced in 2014 to solve the "bottleneck" problem in traditional Encoder-Decoder models used for Machine Learning.

What is Bahdanau Attention?

In traditional sequence-to-sequence (Seq2Seq) models, the encoder compresses the entire input sentence into a single, fixed-length vector (the context vector). Bahdanau Attention allows the model to "look back" at all the source words and focus on specific parts of the input sequence for every word it predicts.

Unlike later mechanisms (like Luong/Dot-Product attention), Bahdanau’s approach calculates the alignment scores using a learned feed-forward network.

When and Where is it Used?

    When: It is primarily used when dealing with long sequences where a single fixed vector cannot capture all the necessary information.

    Where: * Neural Machine Translation (NMT): Translating long sentences between languages.

        Image Captioning: Focusing on specific pixels in an image to describe objects.

        Speech Recognition: Aligning audio frames with text characters.

Why is it Used?

    Overcoming the Bottleneck: Standard RNNs lose information from the beginning of a sentence by the time they reach the end. Attention ensures every input word is accessible.

    Handling Long Sequences: It significantly improves performance on sentences longer than 20–30 words.

    Interpretability: It allows us to visualize which source words the model prioritized when generating a specific output word (via an attention map).

How is it Used? (The Mechanism)

The process follows these mathematical steps:

    Alignment Scores: The model takes the current decoder hidden state st−1​ and all encoder hidden states hi​. It passes them through a small neural network to get a score:
    eti​=va⊤​tanh(Wa​[st−1​;hi​])

    Weights: These scores are normalized using a Softmax function to create a probability distribution (αti​).

    Context Vector: A weighted sum of the encoder hidden states is calculated:
    ct​=∑αti​hi​

    Decoding: This context vector ct​ is combined with the decoder's current state to predict the next word.

Pros and Cons
Pros

    High Accuracy on Long Sequences: Effectively captures long-range dependencies that standard RNNs/LSTMs miss.

    Flexible Alignment: Because it uses a learned weight matrix (Wa​), it can learn complex relationships between the input and output.

    Better Gradient Flow: It provides a "shortcut" for gradients to flow back to earlier parts of the input, reducing the vanishing gradient problem.

Cons

    Computational Expense: Calculating the alignment scores for every single step of the decoder is resource-intensive.

    Speed: It is slower than "Dot-Product" (Luong) attention because it requires a feed-forward neural network layer for every alignment.

    Not Parallelizable: Because it is built on top of RNNs, it cannot take full advantage of modern GPUs like Transformers can.

# Luong Attention Mechanism

What is Luong Attention?

Luong Attention is a technique used in sequence-to-sequence models that allows the decoder to selectively "attend" to specific parts of the encoder's input. Unlike Bahdanau attention, which uses the previous hidden state of the decoder, Luong attention uses the current hidden state of the decoder to calculate attention. It is called "multiplicative" because the alignment score is often calculated using a dot product.

When and Where is it Used?

    When: It is used during the decoding phase of a neural network when the model needs to decide which part of the input sequence is most relevant at the current timestamp.

    Where: * Machine Translation: Converting one language to another (e.g., Google Translate’s early NMT systems).

        Text Summarization: Identifying key sentences or words in a long document to generate a short summary.

        Chatbots: Helping the model maintain context over a conversation.

Why is it Used?

    Efficiency: Luong attention offers several "scoring functions" (dot, general, concat) that are computationally faster than the additive approach.

    Performance: It allows the decoder to "see" the entire input sentence, solving the problem where the encoder forgets the beginning of a long sentence.

    Global vs. Local: Luong introduced the concept of Global Attention (looking at all hidden states) and Local Attention (looking at a small window of hidden states), providing flexibility for different task sizes.

How is it Used? (The Mechanism)

The Luong mechanism follows these specific steps:

    Generate Hidden State: The decoder produces a hidden state ht​ for the current time step.

    Alignment Scores: The model calculates a score between the current decoder state ht​ and all encoder states hˉs​. The most common method is the dot product:
    score(ht​,hˉs​)=ht⊤​hˉs​

    Softmax: These scores are turned into weights (probabilities) using the Softmax function.

    Context Vector: A context vector ct​ is created by taking the weighted average of all encoder hidden states.

    Final Hidden State: The context vector and the current decoder hidden state are combined to produce an "attentional" hidden state, which is then used to predict the actual word.

Pros and Cons
Pros

    Faster Computation: The dot-product calculation is much faster and more memory-efficient than the feed-forward network used in Bahdanau attention.

    Simplified Architecture: It is generally easier to implement and integrate into existing RNN/LSTM frameworks.

    Flexibility: Offers different ways to calculate scores (Dot, General, Concat), allowing developers to tune the complexity based on the available hardware.

    Better for Long Sequences: Like all attention mechanisms, it significantly outperforms basic Encoder-Decoder models on long sentences.

Cons

    Theoretical Limitation: The simple dot-product assumes that the encoder and decoder hidden states live in the same space; if they don't, you must use the "General" score function, which adds a weight matrix and increases complexity.

    Linearity: It may struggle with extremely complex, non-linear relationships between languages compared to the "Additive" (Bahdanau) approach.

    Sequential Bottleneck: Since it is still based on RNNs, it cannot be fully parallelized like a Transformer model.
