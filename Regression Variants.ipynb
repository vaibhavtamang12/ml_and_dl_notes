# Ridge Regression

## What is Ridge Regression?

Ridge Regression, also known as L2 Regularization, is an extension of linear regression that adds a penalty term to the standard loss function. While standard linear regression (Ordinary Least Squares or OLS) tries to minimize only the sum of squared errors, Ridge Regression adds a "penalty" proportional to the square of the magnitude of the coefficients.

Mathematically, the objective function to minimize is:$$J(\beta) = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda \sum_{j=1}^{p} \beta_j^2$$

Where:

$\lambda$ (Lambda): The tuning parameter (complexity parameter) that controls the strength of the penalty.

$\beta_j$: The coefficients (weights) of the model.

When and Why is it Used?

1. Handling Multicollinearity: In standard regression, if two or more predictor variables are highly correlated (e.g., "house square footage" and "number of rooms"), the model becomes unstable. Small changes in the data can lead to wild swings in the coefficients. Ridge regression "shrinks" these coefficients, making them more stable and reliable.

2. Preventing Overfitting: When you have a large number of features ($p$) relative to the number of observations ($n$), OLS tends to overfit, capturing "noise" rather than the actual pattern. Ridge regression constrains the coefficients so the model doesn't rely too heavily on any single feature.

3. The Bias-Variance Tradeoff: Ridge regression deliberately introduces a small amount of bias into the model to significantly reduce its variance. This often leads to a lower Mean Squared Error (MSE) on new, unseen data, making the model generalize better.

How is it Used?

To implement Ridge Regression, you typically follow these steps:

Standardization: Because the penalty is based on the square of the coefficients, it is crucial to scale your features (e.g., using Z-score normalization) so that variables with larger units don't unfairly dominate the penalty term.

Choosing $\lambda$:

- If $\lambda = 0$, the model is identical to OLS.

- As $\lambda \to \infty$, the coefficients $\beta$ shrink toward zero (but never exactly reach it).

Cross-Validation: The "best" $\lambda$ is usually found using techniques like K-Fold Cross-Validation, where you test different values and pick the one that results in the lowest error on a validation set.

Pros of Ridge Regression

- Reduces Model Complexity: By penalizing large coefficients, it prevents the model from becoming overly sensitive to the specific noise in your training data.

- Solves Multicollinearity: It is specifically designed to handle datasets where independent variables are highly correlated, a scenario where standard linear regression usually fails or produces "wild" coefficients.

- Improved Prediction Accuracy: By trading off a little bit of bias for a significant reduction in variance, it generally achieves better performance on new, unseen data compared to OLS.

- Computational Efficiency: Unlike some complex machine learning algorithms, Ridge regression has a closed-form analytical solution, making it very fast to compute even with large datasets.

- Retains All Variables: It is useful when you have a theoretical reason to keep all predictors in the model, as it shrinks them toward zero but never excludes them entirely.

Cons of Ridge Regression

- No Feature Selection: Because it never shrinks coefficients to exactly zero, it cannot be used to "auto-select" the most important variables. You will still end up with a model containing every original feature.

- Model Interpretability: If you have 1,000 features, Ridge will give you 1,000 tiny coefficients. This can make it difficult to explain to stakeholders which specific variables are the primary drivers of the outcome.

- Requires Parameter Tuning: You must carefully choose the value of $\lambda$ (the penalty strength). If $\lambda$ is too high, the model becomes too simple (underfits); if it's too low, you’re back to the problems of OLS.

- Data Scaling is Mandatory: It is highly sensitive to the scale of your input data. If one variable is measured in kilometers and another in millimeters, the penalty will be applied unfairly unless you standardize the data first.

- Introduction of Bias: By definition, Ridge regression coefficients are biased estimates of the "true" population parameters, unlike OLS which is an Unbiased Estimator (under the right conditions).

# Lasso Regression

## What is Lasso Regression

Lasso Regression (Least Absolute Shrinkage and Selection Operator), also known as L1 Regularization, is a type of linear regression that adds a penalty to the model to prevent overfitting. While Ridge regression uses the square of the coefficients, Lasso uses the absolute value of the coefficients.

Mathematically, the cost function for Lasso is:$$J(\beta) = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda \sum_{j=1}^{p} |\beta_j|$$The key characteristic of Lasso is its ability to shrink some coefficients exactly to zero. This effectively removes those features from the model, making Lasso a built-in tool for feature selection.

When and Why is it Used?

1. High-Dimensional DataLasso is the go-to choice when you have a dataset with many features (large $p$) but only a few observations (small $n$). Because it can zero out variables, it simplifies the model and makes it computationally manageable.

2. Automatic Feature SelectionIn many real-world problems, you might have hundreds of potential predictors, but only 5 or 10 actually matter. Lasso automatically identifies and keeps the "important" variables while discarding the noise.

3. Model InterpretabilityBecause Lasso results in a "sparse" model (one with many zeros), the final equation is much easier to explain to non-technical stakeholders. Instead of saying 200 variables affect your outcome, you can point to the 5 that have non-zero coefficients.

How is it Used?

Standardization: Like Ridge, you must scale your features first. Since the penalty is based on the absolute value of the coefficient, variables with larger units will be penalized more heavily if not standardized.

Tuning Lambda ($\lambda$): * If $\lambda = 0$, you are back to standard Ordinary Least Squares (OLS).

- As $\lambda$ increases, more coefficients are pushed to zero.

Cross-Validation: Usually, a technique like LassoCV is used to test many $\lambda$ values and pick the one that minimizes the mean squared error (MSE) on a validation set.

Pros

- Built-in Feature Selection: Automatically identifies and removes irrelevant variables, creating a simpler model.

- Sparse Solutions: Ideal for models where you want a clear, concise list of important predictors.

- Reduces Overfitting: By penalizing the size of coefficients, it helps the model generalize better to new data.

- Handles Irrelevant Features: Unlike Ridge, which keeps everything, Lasso can completely ignore "junk" variables that add no value.

Cons

- Arbitrary Selection in Collinearity: If two variables are highly correlated (e.g., height in inches vs. height in cm), Lasso will randomly pick one and zero out the other. It doesn't "share" the weight between them.

- Constraint on Number of Features: If the number of predictors ($p$) is greater than the number of observations ($n$), Lasso will select at most $n$ variables before it "fills up."

- Bias: Like all regularization methods, Lasso introduces bias into the estimates.

- Inconsistent for Grouped Effects: It may struggle if a group of variables only works well together; Lasso might pick just one from the group and break the relationship.

# ElasticNet Regression

## What is ElaticNet Regression

Elastic Net Regression is a regularized regression method that combines the penalties of both Ridge (L2) and Lasso (L1) regression. It was designed to overcome the specific limitations of each by finding a middle ground between them.Mathematically, the objective function adds both the absolute and squared values of the coefficients to the error:$$J(\beta) = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda_1 \sum_{j=1}^{p} |\beta_j| + \lambda_2 \sum_{j=1}^{p} \beta_j^2$$

## Why and When is it used

1. Handling Correlated Features (Group Effect)Lasso has a weakness: if two variables are highly correlated, it tends to randomly pick one and discard the other. This can be problematic if you need to understand the influence of both. Elastic Net uses the Ridge component to "group" these variables together, shrinking their coefficients more stably as a set.

2. High-Dimensional Data ($p > n$)In situations where you have more features than data points (e.g., genetics or medical imaging), Lasso can only select at most $n$ features. Elastic Net removes this restriction, allowing you to select more features than observations if necessary.

3. The "Best of Both Worlds"It is the go-to model when you aren't sure whether you should use Ridge or Lasso. It provides a "flexible slider" that lets you decide how much feature selection (Lasso) versus coefficient shrinkage (Ridge) you want.

How is it Used?

Elastic Net introduces a second hyperparameter to balance the two methods. In popular libraries like Scikit-Learn, this is handled via:

1. Alpha ($\alpha$): This controls the overall strength of the penalty.

2. L1 Ratio: This is the "slider" between the two penalties.
  - L1 Ratio = 1: Pure Lasso.
  - L1 Ratio = 0: Pure Ridge.
  - L1 Ratio = 0.5: An equal 50/50 mix of both

3. Standardization: As with the other methods, you must standardize your data (scale features to have a mean of 0 and variance of 1) before running the model.

Pros

- Stability: More stable than Lasso when predictors are highly correlated.

- Flexibility: Can perform feature selection (zeroing out junk variables) while still maintaining the predictive power of correlated features.

- Handles Complexity: Ideal for large datasets with many variables (high dimensionality).

- Generalization: Often achieves the lowest test error compared to OLS, Ridge, or Lasso because it is the most general form of linear regularization.

Cons

- Double Shrinkage Bias: Because it applies two different penalties, the final coefficients can be "shrunken twice," which can lead to higher bias and less accurate predictions if not carefully tuned.

- Harder to Tune: You now have two hyperparameters to optimize ($\alpha$ and $L1\_ratio$) instead of just one ($\lambda$), which requires more computational time and cross-validation.

- Computational Cost: It is slightly more complex to calculate than pure Ridge or Lasso, which might be noticeable on extremely large datasets.
