# Activation Function

An activation function is a mathematical gate attached to each neuron in a neural network. It determines whether a neuron should "fire" (pass information forward) based on the input it receives.

What is an activation function

In deep learning, an activation function is a scalar-to-scalar function applied to the output of a neuron's linear transformation.Mathematically, a neuron calculates the weighted sum of its inputs plus a bias:$$z = \sum (w_i \cdot x_i) + b$$The activation function $f(z)$ is then applied to this sum to produce the final output of that neuron.

Why is it used

The primary purpose of an activation function is to introduce non-linearity into the network.

- The Linear Limitation: Without an activation function, no matter how many layers you add to a neural network, the entire model would still behave like a single-layer linear regression. This is because the composition of linear functions is always a linear function.

- Complex Pattern Recognition: Real-world data (images, audio, financial trends) is rarely linear. Activation functions allow the network to learn complex, high-dimensional patterns and "curved" decision boundaries.

When and where is it used

Activation functions are used during the forward pass of the training and inference phases. They are placed:

- In Hidden Layers: To help the model learn complex features and representations of the data.

- In the Output Layer: To map the model’s final predictions into a specific format (e.g., a probability between 0 and 1, or a range for regression).

How is it used?

The process follows these steps:

- Input: The neuron receives multiple signals ($x$) from the previous layer.

- Weighted Sum: The signals are multiplied by weights ($w$) and a bias ($b$) is added.

- Activation: This sum is passed through the activation function.

- Output: The result is sent to the neurons in the next layer.

Pros

- Non-Linearity: This is their most critical "pro." By introducing non-linear transformations, they allow a neural network to act as a Universal Function Approximator, meaning the model can learn and represent almost any complex relationship between inputs and outputs.

- Feature Selection: They act as a digital gate, helping the network decide which information is "noise" and which is relevant. A neuron with a low activation effectively tells the next layer to ignore its input.

- Gradient Flow Control: Many functions are designed to have mathematical properties (differentiability) that make it easier for the Backpropagation algorithm to calculate how much to change weights during training.

- Output Normalization: They can "squash" or scale the output of a neuron into a specific range (like 0 to 1 or -1 to 1), which is essential for tasks like predicting probabilities or normalizing data for the next layer.

Cons

- Vanishing Gradient Problem: This is a major drawback for certain functions. When the input becomes very large or very small, the gradient (slope) of the function becomes nearly zero. During training, this causes the weight updates to become so tiny that the network stops learning.

- Exploding Gradient Problem: Conversely, some functions can cause gradients to accumulate and become extremely large, leading to unstable weight updates that "break" the model's progress.

- Computational Expense: Functions involving complex math (like exponents or divisions) require more processing power. In a network with millions of neurons, this can significantly slow down training and inference time.

- Dead Neurons (Dying ReLU): Some functions can cause a neuron to output zero for all possible inputs. Once a neuron is "dead" in this way, it no longer contributes to the network's learning process, effectively wasting that part of the model's capacity.

Different types of Activation Function

Sigmoid Function

- The Sigmoid function takes any real value and "squashes" it into a range between 0 and 1.

- Formula: $\sigma(z) = \frac{1}{1 + e^{-z}}$

- Where/When to use: Historically used in hidden layers, but now almost exclusively used in the output layer for binary classification (Yes/No).

- Pros: Clear probabilistic interpretation; smooth gradient.

- Cons: Vanishing Gradient Problem (the ends are flat, so the slope is near zero); outputs are not zero-centered, which can slow down optimization.

TanH

- Tanh is similar to Sigmoid but maps the input to a range between -1 and 1.

- Formula: $tanh(z) = \frac{e^z - e^{-z}}{e^z + e^{-z}}$

- Where/When to use: Preferred over Sigmoid for hidden layers in many older architectures.

- Pros: Zero-centered, meaning negative inputs are mapped strongly negative and positive inputs strongly positive. This helps the model converge faster during training.

- Cons: Still suffers from the Vanishing Gradient Problem when inputs are very high or low.

ReLU

- The current "gold standard" for deep learning. It outputs the input directly if it is positive; otherwise, it outputs zero.

- Formula: $f(z) = \max(0, z)$

- Where/When to use: The default choice for hidden layers in almost all Deep Neural Networks (CNNs, MLPs).

- Pros: Extremely computationally efficient (just a threshold check); avoids the vanishing gradient for positive values, allowing for very deep models.

- Cons: Dying ReLU Problem—if a neuron’s weights are updated such that it always outputs 0, it becomes "dead" and never learns again.

Leaky ReLU

- A variant of ReLU designed to fix the "dying" problem by allowing a small, non-zero gradient when the input is negative.

- Formula: $f(z) = \max(\alpha z, z)$, where $\alpha$ is a small constant like 0.01.

- Where/When to use: When you notice a high percentage of dead neurons in your ReLU-based model.

- Pros: Prevents dead neurons by ensuring the gradient is never exactly zero.

- Cons: Results can be inconsistent depending on the value of $\alpha$; slightly more computation than standard ReLU.

Softmax

- Softmax is a unique function that turns a vector of numbers into a vector of probabilities that sum up to 1.

- Formula: $\sigma(\mathbf{z})_i = \frac{e^{z_i}}{\sum e^{z_j}}$

- Where/When to use: Almost strictly used in the final output layer for multi-class classification (e.g., identifying if an image is a cat, dog, or bird).

- Pros: Provides a clear probability distribution across multiple categories.

- Cons: Can be computationally expensive for a very large number of classes (e.g., millions of words in language models).
