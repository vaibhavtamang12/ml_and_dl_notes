# Batch Normalization



In deep learning, training deep neural networks can be a finicky process. As parameters change in one layer, the distribution of inputs to the next layer changes—a phenomenon known as Internal Covariate Shift.

Batch Normalization (BatchNorm) is the industry-standard "stabilizer" that keeps these distributions consistent, making training faster and less sensitive to how you initialize your weights.

What is Batch Normalization

Batch Normalization is a technique that normalizes the contributions to a layer for every mini-batch. It shifts the mean to 0 and the variance to 1. However, because a simple normalization might limit what the layer can learn, BatchNorm introduces two learnable parameters: Gamma ($\gamma$) for scaling and Beta ($\beta$) for shifting.

The transformation follows these steps for a mini-batch $B$:

- Calculate Mean: $\mu_B = \frac{1}{m} \sum_{i=1}^{m} x_i$

- Calculate Variance: $\sigma^2_B = \frac{1}{m} \sum_{i=1}^{m} (x_i - \mu_B)^2$

- Normalize: $\hat{x}_i = \frac{x_i - \mu_B}{\sqrt{\sigma^2_B + \epsilon}}$ (where $\epsilon$ is a tiny constant for numerical stability)

- Scale and Shift: $y_i = \gamma \hat{x}_i + \beta$

Why do we use it

Without BatchNorm, small changes in early layers amplify as they propagate deeper. This forces later layers to constantly adapt to a "moving target." By normalizing the inputs, we ensure:

- Faster Convergence: You can use higher learning rates without the model exploding or vanishing.

- Weight Independence: The network is less sensitive to the initial values of the weights.

- Gradient Flow: It keeps activations in a regime (like the non-saturated region of a Sigmoid or Tanh) where gradients stay healthy.

When and where do we use it

When: Almost always in deep Convolutional Neural Networks (CNNs) and deep Feedforward networks.

Where: It is typically placed after the convolutional or fully connected layer and before the activation function (like ReLU), though some researchers experiment with placing it after the activation.

Pros and Cons

Pros

- Increases Training Speed: By stabilizing the learning process, the model reaches the optimum much faster.

- Acts as Regularization: Because the mean and variance are calculated on a per-batch basis, it adds a slight "noise" to the data, which can reduce overfitting (similar to Dropout).

- Reduces Sensitivity to Initialization: You don't have to spend hours tuning your weight initialization strategy.

- Allows Higher Learning Rates: You can be more aggressive with your optimizer.

Cons

- Dependency on Batch Size: It doesn't work well with very small batch sizes (e.g., batch size of 2 or 4), as the mean/variance estimates become inaccurate.

- Inconsistent between Train and Test: During inference (testing), you don't have a "batch," so you have to use a running average of the mean and variance saved during training, which can sometimes lead to discrepancies.

- Memory Overhead: It adds extra parameters ($\gamma$ and $\beta$) and computation time to each forward/backward pass.

- Harder for RNNs: BatchNorm is notoriously difficult to implement in Recurrent Neural Networks (where Layer Normalization is usually preferred instead).
