# AdamW

AdamW is a sophisticated modification of the Adam optimizer that fixes a specific mathematical flaw in how weight decay (regularization) is handled. It has become the gold standard for training modern deep learning models, particularly Transformers.

When and where is it used

AdamW is the "go-to" optimizer for state-of-the-art (SOTA) models. You will see it used almost exclusively in:

- Large Language Models (LLMs): Training models like GPT-4, Llama, and BERT.

- Vision Transformers (ViT): Modern image recognition models that replace traditional CNNs.

- Transfer Learning: When fine-tuning a pre-trained model on a new, smaller dataset.

Why is it used

The "W" stands for Weight Decay. In the original Adam optimizer, L2 regularization (which keeps weights small to prevent overfitting) was mathematically combined with the gradient updates.

However, because Adam uses adaptive learning rates for each parameter, the L2 penalty was unintentionally scaled by the moving averages. This meant that parameters with large gradients were regularized less than those with small gradients. AdamW decouples the weight decay from the gradient update, ensuring that regularization is applied consistently across all weights.

How is it used

In libraries like PyTorch or Hugging Face, you simply replace torch.optim.Adam with torch.optim.AdamW.

Mathematically, instead of adding the penalty to the gradient ($g_t$), AdamW subtracts a fraction of the weight directly during the update step:$$w_{t+1} = w_t - \eta \left( \frac{\hat{m}_t}{\sqrt{\hat{v}_t} + \epsilon} + \lambda w_t \right)$$

- $\lambda$ is the weight decay coefficient.

- The term $\lambda w_t$ is applied separately from the adaptive gradient component.

Pros and Cons

Pros

- Better Generalization: By fixing the weight decay logic, AdamW consistently achieves lower test error and generalizes better to new data than standard Adam.

- Standard for Transformers: It is the most stable optimizer for high-parameter models like Transformers and LLMs.

- Stable Training: It prevents weights from growing too large without interfering with the "learning" direction of the gradients.

- Reliable Hyperparameters: The default settings for $\beta_1$ and $\beta_2$ usually work very well across diverse tasks.

Cons

- Extra Hyperparameter: You must carefully tune the weight decay coefficient ($\lambda$), as it has a significant impact on performance.

- Memory Usage: Like Adam, it requires storing two moving averages for every single parameter in the network, which can be memory-intensive for massive models.

- Sensitive to Learning Rate: While adaptive, the performance of AdamW can still fluctuate significantly if the global learning rate is not scheduled correctly (e.g., using a linear warmup).
