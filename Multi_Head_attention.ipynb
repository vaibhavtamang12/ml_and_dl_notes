# Multi Head Attention

Multi-head attention (MHA) is the core mechanism that allows Transformer models to process information in parallel across different "representation subspaces." Instead of looking at a sentence or a piece of data through a single lens, the model splits its focus into several independent heads that can attend to different parts of the input simultaneously.

What is Multi-Head Attention?

In a standard attention mechanism, the model calculates a single set of attention scores for each word. Multi-head attention improves this by linearly projecting the input (Queries, Keys, and Values) h times with different, learned weight matrices. Each "head" then performs the Scaled Dot-Product Attention independently. The results are concatenated and transformed back to the original dimension.

The formula for each head i is:

For the individual head calculation:
$$\text{head}_i = \text{Attention}(QW_i^Q, KW_i^K, VW_i^V)$$

For the final multi-head output:
$$\text{MultiHead}(Q, K, V) = \text{Concat}(\text{head}_1, \dots, \text{head}_h)W^O$$

When and Where is it Used?

Multi-head attention is the primary engine of the Transformer architecture. It is used in:

    Encoder layers: To help the model understand the context of a word within the entire input sequence.

    Decoder layers (Self-attention): To allow the model to focus on previous words in the generated sequence while masking future words.

    Encoder-Decoder attention: To let the decoder "look" at the encoder's output, essentially focusing on relevant parts of the input while generating the output (vital for translation).

Why is it Used?

If you use only one attention head, the model tends to average out all the information. For example, in the sentence "The animal didn't cross the street because it was too tired," a single head might struggle to decide if "it" refers to the animal or the street.

MHA allows the model to:

    Capture multiple types of relationships: One head can focus on grammar, another on coreference (like the "it" example), and another on the verb-object relationship.

    Enhance Representational Power: It prevents the model from being "blind" to certain parts of the sequence by providing multiple points of view.

How is it Used?

    Linear Projection: The input is multiplied by different weight matrices to create Q, K, and V for each head.

    Parallel Attention: Each head calculates attention scores.

    Concatenation: The outputs from all heads are merged.

    Final Linear Layer: A final weight matrix WO ensures the output shape matches the input, allowing for residual connections.

Pros and Cons

Pros

    Parallelism: Different heads can be computed at the same time, making it highly efficient on modern hardware like GPUs.

    Rich Feature Extraction: It allows the model to learn complex, nuanced relationships in the data that a single attention mechanism would miss.

    Stability: It acts similarly to an "ensemble" of attention mechanisms, which often leads to more stable and accurate learning.

    Global Context: Unlike RNNs, it can connect distant words in a sequence regardless of their position.

Cons

    High Memory Usage: The memory cost scales quadratically O(n2) with the sequence length (n), making it expensive for very long documents.

    Computational Complexity: Calculating multiple projections and attention scores for every head adds significant floating-point operations.

    Redundancy: Research shows that some heads often learn the same things or become "dead," meaning the model might be carrying more parameters than it strictly needs.

    Requires Large Datasets: Because of the high number of parameters and complex structure, MHA-based models usually require massive amounts of data to train effectively.
