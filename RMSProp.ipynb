# RMSProp

RMSProp (Root Mean Square Propagation) is an adaptive learning rate optimization algorithm designed to solve the primary flaw of AdaGrad: the fact that the learning rate eventually shrinks to zero, causing the model to stop learning.

What is RMSProp

RMSProp modifies AdaGrad by using a moving average of squared gradients instead of a cumulative sum. This means the algorithm "forgets" very old gradients and focuses on the most recent ones, preventing the denominator from growing indefinitely.

The Mathematical Logic

The update rule for a parameter $\theta$ at time $t$ is:

- Calculate the moving average:$$E[g^2]_t = \beta E[g^2]_{t-1} + (1 - \beta)g_t^2$$

- Update the parameters:$$\theta_{t+1} = \theta_t - \frac{\eta}{\sqrt{E[g^2]_t + \epsilon}} \cdot g_t$$

Where:

- $E[g^2]_t$: The exponentially decaying average of squared gradients.

- $\beta$: The forgetting factor (usually set to $0.9$).

- $\eta$: The learning rate.$g_t$: The current gradient.

- $\epsilon$: A small constant to prevent division by zero.

When and where is it used

RMSProp is a go-to optimizer for many deep learning tasks, particularly:

- Recurrent Neural Networks (RNNs): It is exceptionally effective at handling the vanishing or exploding gradient problems common in sequence modeling.

- Mini-batch Learning: It performs well when gradients are calculated over small subsets of data rather than the whole dataset.

- Non-Stationary Objectives: In environments where the "best" direction changes over time (like in Reinforcement Learning), RMSProp’s ability to "forget" old gradients is vital.

Why is it used?

The main reason for using RMSProp is to maintain a stable learning speed.

- The Problem with AdaGrad: In AdaGrad, the denominator keeps growing until the learning rate effectively hits zero. This is like a car that slowly runs out of gas and stops before reaching its destination.

- The RMSProp Solution: By using a weighted moving average, the "gas tank" (learning rate) stays replenished based on recent terrain. This allows the model to continue learning and navigating complex loss landscapes indefinitely.

How is it used ?

RMSProp was famously introduced in a Coursera lecture by Geoffrey Hinton rather than a formal paper. It is now a standard tool in all major AI libraries:

- PyTorch: optimizer = torch.optim.RMSprop(model.parameters(), lr=0.01, alpha=0.99)

- TensorFlow: optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.001, rho=0.9)

Pros and Cons

Pros

- Prevents Learning Stagnation: Unlike AdaGrad, it doesn't let the learning rate vanish, allowing the model to converge even in very deep networks.

- Faster Convergence: It adjusts the step size for each parameter, meaning it can speed up in flat areas and slow down in steep areas.

- Excellent for RNNs: It handles the complex, oscillating gradients of sequential data better than most basic optimizers.

- Simple to Implement: It requires very little extra memory compared to more complex algorithms like Adam.

Cons

- Hyperparameter Dependent: You still have to choose a good initial learning rate ($\eta$) and a decay rate ($\beta$).

- Lack of Momentum: Vanilla RMSProp doesn't include a momentum term (though it can be added), which means it might still struggle to escape some local minima compared to NAG or Adam.

- Global Learning Rate: While it scales individual parameters, the "base" learning rate is still a global constant that needs tuning.
