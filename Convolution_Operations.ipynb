# Convolution Operations

In Convolutional Neural Networks (CNNs), a convolution operation is a mathematical process used to extract features from an input (like an image). Unlike traditional neural networks that treat every pixel as an independent variable, convolutions look at pixels in context with their neighbors.

How they work

The operation follows a simple mathematical process called a dot product:

- Placement: The filter is placed over a patch of the input image starting at the top-left corner.

- Multiplication: Each value in the filter is multiplied by the corresponding pixel value it is currently "covering."

- Summation: All these individual products are summed together to produce a single number.

- Sliding: The filter moves (strides) to the next position, and the process repeats until the entire image is covered.

Mathematically, for an input $I$ and a kernel $K$, the value at position $(i, j)$ of the output is:$$S(i, j) = \sum_{m} \sum_{n} I(i+m, j+n) \cdot K(m, n)$$

Key Hyperparams

The behavior of the convolution is controlled by a few settings:

- Stride: The number of pixels the filter shifts at each step. A stride of 1 moves one pixel at a time; a stride of 2 skips a pixel, effectively shrinking the output size.

- Padding: Adding "fake" pixels (usually zeros) around the border of the input. This allows the filter to cover the edges of the image and ensures the output dimensions don't shrink too rapidly.

- Channels: In a color image, the convolution happens across three layers (Red, Green, Blue) simultaneously, and the results are summed into a single feature map.

- Dilation: Adding gaps between kernel elements to increase the "receptive field" without increasing the number of parameters.
