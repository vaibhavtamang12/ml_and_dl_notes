# Nesterov Accelerated Gradient

Nesterov Accelerated Gradient (NAG) is an optimized version of the standard Momentum optimization technique. It is designed to help gradient descent converge faster and more accurately by "looking ahead" before making a step.

What is NAG?

In standard Momentum, we calculate the gradient at our current position and then add the momentum from previous steps. In NAG, we do the opposite: we first jump in the direction of the momentum and then calculate the gradient at that new "look-ahead" position to make a correction.

The Mathematical Logic

The update rule can be expressed as:$$v_t = \gamma v_{t-1} + \eta \nabla_\theta J(\theta - \gamma v_{t-1})$$$$\theta = \theta - v_t$$

Where:

- $\theta$: Parameters being optimized.

- $\gamma$: Momentum coefficient (usually around 0.9).

- $\eta$: Learning rate.

- $\nabla_\theta J(\theta - \gamma v_{t-1})$: The gradient calculated at the "ahead" position.

When and Where is it Used?

NAG is primarily used in Deep Learning and Neural Network training. It is a standard feature in most optimization libraries (like PyTorch or TensorFlow) and is often used when:

- Training deep architectures where standard SGD is too slow.

- The loss landscape is complex, with many "plateaus" or "saddles" that require momentum to escape.

- You are looking for a more stable alternative to standard Momentum.

Why is it used ?

The main reason for using NAG is to prevent overshooting.

Imagine a ball rolling down a hill into a valley. With standard momentum, the ball gains so much speed that it flies past the bottom and has to roll back. NAG acts like a "smart" brake; because it calculates the gradient at the predicted next position, it senses if it is about to go back up the other side of the hill and slows down the update accordingly.

How is it Used?

In practice, you rarely implement the calculus from scratch. Most modern frameworks allow you to enable it with a simple flag within the SGD optimizer.

PyTorch: optimizer = torch.optim.SGD(model.parameters(), lr=0.01, momentum=0.9, nesterov=True)

TensorFlow/Keras: optimizer = tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9, nesterov=True)

Pros and Cons

Pros

- Faster Convergence: It generally reaches the minimum faster than standard Gradient Descent or vanilla Momentum.

- Reduced Oscillations: By "looking ahead," it dampens the zig-zagging motion common in optimization, leading to a smoother path.

- Greater Stability: It is less likely to diverge or overreact to sharp changes in the gradient.

- Efficiency: It provides many of the benefits of complex adaptive methods (like Adam) but with lower computational overhead.

Cons

- Manual Tuning: Like standard SGD, it still requires careful tuning of the learning rate and the momentum coefficient ($\gamma$).

- Hyperparameter Sensitivity: If the momentum is set too high, the "look-ahead" might lead to inaccurate corrections.

- Not Adaptive: Unlike Adam or RMSProp, it does not adjust the learning rate for each individual parameter; it applies a global rate.
