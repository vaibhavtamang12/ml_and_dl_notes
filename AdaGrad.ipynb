# AdaGrad

AdaGrad (Adaptive Gradient Algorithm) is an optimization method that scales the learning rate for each individual parameter based on the history of gradients. While standard Gradient Descent uses a single learning rate for all weights, AdaGrad gives "custom" rates to every feature.

What is Adagrad

The core idea is adaptation. It performs larger updates for infrequent parameters and smaller updates for frequent parameters. It does this by keeping a running sum of the squares of all past gradients for each parameter.

The Mathematical Logic

The update rule for a parameter $\theta_i$ at time step $t$ is:$$\theta_{t+1, i} = \theta_{t, i} - \frac{\eta}{\sqrt{G_{t, ii} + \epsilon}} \cdot g_{t, i}$$

Where:

- $\eta$: The initial global learning rate.

- $g_{t, i}$: The gradient of the loss function for parameter $i$ at time $t$.

- $G_{t, ii}$: A diagonal matrix where each diagonal element $i,i$ is the sum of the squares of the gradients for $\theta_i$ up to time $t$.

- $\epsilon$: A small smoothing term (usually $10^{-8}$) to avoid division by zero.

When and Where is it Used?

AdaGrad excels in scenarios involving sparse data. It is commonly found in:

- Natural Language Processing (NLP): Where some words (features) appear very rarely while others appear constantly.

- Recommendation Systems: Where user-item interactions are sparse (most users only interact with a tiny fraction of items).

- Computer Vision: Specifically in the early layers of some older architectures to handle varying feature frequencies.

Why is it used?

In many real-world datasets, features are not distributed equally.

- The Problem: If you use a single learning rate, rare features might never be updated enough to be "learned," while common features might cause the model to over-correct.

- The AdaGrad Solution: Because it divides the learning rate by the square root of accumulated gradients, parameters that receive frequent, large updates see their learning rate decrease rapidly. Parameters that are rarely updated maintain a higher effective learning rate, allowing them to "catch up."

How is it used

Most machine learning frameworks implement AdaGrad as a standard optimizer. You typically only need to set the initial learning rate.

- PyTorch: optimizer = torch.optim.Adagrad(model.parameters(), lr=0.01)

- TensorFlow: optimizer = tf.keras.optimizers.Adagrad(learning_rate=0.01)

Pros and Cons

Pros

- Automatic Learning Rate Tuning: It eliminates the need to manually tune the learning rate as strictly as standard SGD.

- Handles Sparse Data: It is arguably one of the best choices for datasets where features have vastly different frequencies.

- Increased Robustness: It is less sensitive to the initial learning rate than vanilla Gradient Descent.

Cons

- Vanishing Learning Rate: This is its biggest weakness. Because the accumulated squared gradients ($G_{t}$) only ever increase, the effective learning rate eventually shrinks to an infinitesimally small value, causing the model to stop learning entirely before reaching the minimum.

- Memory Intensive: It must store the sum of squared gradients for every single parameter in the model.

- Global Rate Sensitivity: While it adapts, a poor choice for the initial global learning rate ($\eta$) can still cause the algorithm to "die" too early
