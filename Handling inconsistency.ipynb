Handling Missing Data

We have two approaches to handle missing values:

- Remove them

- Impute

For imputing, we have imputing of univariate and multivariate.

For univariate numerical we can use:

- Mean/Median
- Random Initialization
- End of Distribution

For univariate categorical we can use:

- Mode
- Fill with word "missing"

For multivariate imputation we can use:

- KNN Imputer
- Iterative Imputer (MICE)

### Mean / Median Imputation

When to use: Data is missing completely at random (MCAR). Use Mean if the distribution is Gaussian (Normal); use Median if there are outliers or the data is skewed.

Pros: Simple to implement; maintains the sample size.

Cons: Distorts the variance and covariance of the dataset; ignores relationships between features

Random Sample Imputation

When to use: When you want to preserve the original distribution of the data.

Pros: Maintains the variance and shape of the histogram.

Cons: Introduces randomness; two different runs might yield different model results.

End of Distribution Imputation

When to use: When data is Missing Not At Random (MNAR). It flags the "missingness" by placing a value far outside the normal range (e.g., Mean + 3$\sigma$).

Pros: Automatically captures the importance of the value being missing.

Cons: Can create outliers that negatively affect linear models.

Mode Imputation

When to use: Categorical data with low missingness (usually < 5%).

Pros: Simple and fast.

Cons: If missingness is high, it over-represents the most frequent category, leading to biased results.

"Missing" Category Imputation

When to use: When the fact that data is missing is itself a signal (e.g., a "Credit Card Type" being missing might mean the user doesn't have one).

Pros: Does not lose any information; very robust.

Cons: Adds a new category, which increases complexity in one-hot encoding.

KNN Imputer

When to use: Small to medium datasets where local similarity matters. It finds the "$k$" most similar rows (neighbors) and averages their values.

Pros: More accurate than univariate methods; captures local correlations.

Cons: Very computationally expensive (slow on large datasets); sensitive to outliers.


Iterative Imputer (MICE)

When to use: Large datasets where features are correlated. It models each feature with missing values as a function of other features in a round-robin fashion.

Pros: Very powerful and flexible; considers the entire dataset's relationships.

Cons: Can be slow; assumes missingness can be modeled linearly (unless using tree-based estimators).

## Handling Outliers

We can handle outliers using three methods:

- Using z-score

- Using Percentile

- Using IQR Method

### Z-Score Method

The Z-score represents how many standard deviations ($\sigma$) a data point is from the mean ($\mu$).

How it works:You calculate the Z-score for every data point using the formula:$$Z = \frac{x - \mu}{\sigma}$$

Generally, a threshold of 3 or -3 is used. If a Z-score is greater than 3 or less than -3, the point is considered an outlier.

When to use: Use this when your data follows a Normal (Gaussian) Distribution.

Pros: Very precise for bell-curve data; easy to interpret in terms of probability.

Cons: Not effective for skewed data; the mean and standard deviation themselves are sensitive to outliers, which can "leak" the outlier's influence into the Z-score calculation.

IQR (Interquartile Range) Method

The IQR method uses the "Middle 50%" of the data to define boundaries. It is the foundation for the "whiskers" on a Box Plot.

How it works:

- Calculate Q1 (25th percentile) and Q3 (75th percentile).

- Calculate $IQR = Q3 - Q1$.

- Define the Lower Limit: $Q1 - (1.5 \times IQR)$.

- Define the Upper Limit: $Q3 + (1.5 \times IQR)$.Anything outside these limits is an outlier.

When to use: Use this for Skewed data or when you do not want to assume a normal distribution.

Pros: Extremely robust; unlike the Z-score, the IQR is not influenced by the outliers it is trying to detect.

Cons: It might be too aggressive for some datasets, labeling valid extreme values as outliers.

Percentile Method (Capping/Winsorization)

This method involves cutting off a fixed percentage of data from the top and bottom ends of the distribution.

How it works:

You pick a threshold (e.g., 1% and 99%). Any data point below the 1st percentile or above the 99th percentile is either removed or "capped" (replaced with the threshold value).

When to use: Use this when you have a very large dataset or when you want to handle "heavy-tailed" distributions where you know the extreme 1% is likely noise or errors.

Pros: Very flexible; you can decide exactly how much data you are willing to lose or modify.

Cons: Arbitrary; there is no statistical "rule" for whether you should use the 1st, 5th, or 10th percentile—it depends entirely on your judgment.
