# Padding and Strides

In deep learning, specifically within Convolutional Neural Networks (CNNs), padding and strides are fundamental hyperparameters that determine how a filter (or kernel) moves across an image and what the resulting output shape will be.

What are padding and strides

Padding

Padding involves adding extra pixels (usually zeros) around the border of an input image before applying the convolution operation.

    "Valid" Padding: No padding is added. The output size shrinks.

    "Same" Padding: Enough padding is added so that the output has the same spatial dimensions as the input.

Strides

Stride is the "step size" the filter takes as it moves across the image.

    A Stride of 1 means the filter moves one pixel at a time.

    A Stride of 2 means the filter jumps two pixels, effectively skipping every other location.

When, where and why are they used

Where: They are used in the Convolutional Layers and Pooling Layers of a neural network.

When: You configure these during the architecture design phase of a model (e.g., in TensorFlow/Keras or PyTorch).

Why: * Padding is used to prevent the loss of information at the edges of an image and to control the shrinking of data as it passes through many layers.

    Strides are used to reduce the spatial dimensions (downsampling), which helps in reducing computational load and increasing the "receptive field" (allowing the model to see a larger area of the image at once).

How are they used

The output dimensions ($H_{out}, W_{out}$) are calculated using the following formula:$$H_{out} = \left\lfloor \frac{H_{in} + 2P - K}{S} \right\rfloor + 1$$

Where:

- $H_{in}$ = Input height

- $P$ = Padding

- $K$ = Kernel (filter) size

- $S$ = Stride

Pros and Cons

Padding

    Pros:

        Preserves Edge Information: Without padding, pixels at the corners are only "seen" once by the filter, while center pixels are seen multiple times. Padding gives edge pixels more "weight."

        Allows for Deeper Networks: By maintaining the spatial size (Same Padding), you can stack dozens of layers without the image shrinking to 1×1 too quickly.

    Cons:

        Increased Computation: Adding extra pixels means the filter has more positions to visit, slightly increasing processing time.

        Artificial Data: Adding zeros introduces "fake" information at the borders, which can sometimes introduce artifacts in specific tasks like image generation.

Strides

    Pros:

        Dimensionality Reduction: Higher strides quickly reduce the size of the feature maps, making the model faster and less memory-intensive.

        Larger Receptive Field: Moving in larger steps helps the deeper layers of the network capture more global features (like the shape of a face) rather than just local ones (like an edge).

    Cons:

        Information Loss: Because you are skipping pixels, you might lose fine-grained spatial details.

        Less Overlap: Higher strides reduce the overlap between filter applications, which can lead to less robust feature detection.
