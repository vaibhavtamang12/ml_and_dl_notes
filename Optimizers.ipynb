# Optimizers

An optimizer is a mathematical algorithm or function used in deep learning to adjust the attributes of a neural network, such as weights and learning rate, to reduce the losses.

Think of it as the "steering wheel" of the training process. While the loss function tells you how far off your predictions are, the optimizer tells the model exactly how to change its parameters to improve those predictions.

Where and when is it used

Where: It resides within the training loop of a machine learning framework (like TensorFlow, PyTorch, or JAX). It sits between the calculated loss and the model’s weights.

When: It is used during the Backpropagation phase of training. After the model makes a "Forward Pass" and calculates the error, the optimizer kicks in to update the weights before the next cycle begins.

Why is it used

Deep learning models often have millions of parameters. Manually finding the perfect combination of weights to minimize error is impossible. The optimizer is used to:

- Minimize the Loss Function: To reach the "global minimum" where the model's error is at its lowest.

- Increase Speed: To help the model learn faster by taking efficient steps toward the solution.

- Improve Accuracy: To ensure the model doesn't just memorize data but actually finds the best patterns.

How is it used?

The process follows a specific cycle:

- Calculate Gradients: The model calculates the derivative (gradient) of the loss function with respect to each weight.

- Determine Direction: The optimizer looks at the gradient to see which direction "downhill" is.

- Update Weights: It subtracts a portion of the gradient from the current weight. The size of this step is determined by the Learning Rate ($\eta$).

The fundamental update formula for Gradient Descent is:$$w_{new} = w_{old} - \eta \cdot \nabla L(w_{old})$$

Pros and Cons

Pros

- Automation: They eliminate the need for manual weight tuning, making complex AI possible.

- Adaptability: Modern optimizers (like Adam) automatically adjust the learning rate for each parameter, which handles sparse data well.

- Efficiency: They use "Momentum" to help the model sail past small "dips" (local minima) in the error surface that might otherwise trap it.

- Convergence: They provide a systematic way to ensure a model eventually stops learning once it has reached its best state.

Cons

- Sensitivity to Hyperparameters: If you set the learning rate too high, the optimizer might "overshoot" the solution; too low, and it may take forever to train.

- Computational Cost: Some advanced optimizers require more memory and processing power because they track "moving averages" of previous steps.

- Risk of Local Minima: Simpler optimizers can get "stuck" in a sub-optimal spot, thinking they’ve found the best solution when a better one exists further away.

- Overfitting: If an optimizer is too aggressive, it might help the model learn the "noise" in the training data rather than the actual patterns.

Different types of optimizers

1. Gradient Descent Variants
Batch Gradient Descent: Uses the entire dataset to update weights once.

Stochastic Gradient Descent (SGD): Updates weights for every single training example.

Mini-Batch Gradient Descent: The industry standard; updates weights based on small groups (batches) of data.

2. Adaptive Learning Rate Optimizers
These are popular because they automatically tune the learning rate during training:

Adagrad (Adaptive Gradient): Adjusts the learning rate based on how frequently a parameter is updated.

RMSProp (Root Mean Square Propagation): Developed to resolve Adagrad's radically diminishing learning rates.

Adam (Adaptive Moment Estimation): Currently the most widely used; it combines the benefits of RMSProp and Momentum.

AdaDelta: An extension of Adagrad that seeks to reduce its aggressive, monotonically decreasing learning rate.

3. Advanced & Niche Optimizers
AdamW: A version of Adam with improved weight decay (common in Transformer models).

Nadam: Combines Adam with Nesterov Accelerated Gradient (NAG).

L-BFGS: A second-order optimization algorithm often used for smaller datasets or specific tasks like Neural Style Transfer
